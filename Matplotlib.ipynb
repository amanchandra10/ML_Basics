```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","y","g","b","r"]
plt.xlabel("Language",fontsize=15)
plt.ylabel("City",fontsize=15)
plt.title("Data")
plt.bar(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","y","g","b","r"]
sizes=[120,223,456,765,111]
plt.xlabel("Language",fontsize=15)
plt.ylabel("City",fontsize=15)
plt.title("Data")
plt.scatter(x,y,color=c,marker="*",s=sizes,edgecolor="b")
plt.show()
```

```python
import matplotlib.pyplot as plt 
y=["c","c++","java","python"]
x=[10,20,30,40]
colors=["r","b","g","y"]
explode=[0.15,0.0,0.0,0.0]
plt.pie(x,labels=y,radius=1.5, explode=explode,shadow=True,autopct="%0.2f%%",colors=colors)
plt.show()
```

```python
#Stack/Area Plot
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
colors=["r","b","g","y"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=colors)
plt.title("Data")
plt.xlabel("x-axis")
plt.ylabel("y-axix")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
colors=[10,20,30,40,50,60,70]
sizes=(500,300,600,300,500,700,800)
plt.title("Data")
plt.xlabel("x-axis")
plt.ylabel("y-axix") 
scatter= plt.scatter(x,y,c=colors,marker="*",s=sizes,cmap="viridis")
colorbar= plt.colorbar(scatter)
colorbar.set_label("colorbar",fontsize=15)  
plt.show()
```
