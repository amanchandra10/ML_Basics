```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","y","g","b","r"]
plt.xlabel("Language",fontsize=15)
plt.ylabel("City",fontsize=15)
plt.title("Data")
plt.bar(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=["r","y","g","b","r"]
sizes=[120,223,456,765,111]
plt.xlabel("Language",fontsize=15)
plt.ylabel("City",fontsize=15)
plt.title("Data")
plt.scatter(x,y,color=c,marker="*",s=sizes,edgecolor="b")
plt.show()
```

```python
import matplotlib.pyplot as plt 
y=["c","c++","java","python"]
x=[10,20,30,40]
colors=["r","b","g","y"]
explode=[0.15,0.0,0.0,0.0]
plt.pie(x,labels=y,radius=1.5, explode=explode,shadow=True,autopct="%0.2f%%",colors=colors)
plt.show()
```
